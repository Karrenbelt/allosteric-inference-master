```python
import os, settings
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
df = pd.read_excel(settings.DATA_DIR+'/Link_2014_S2.xlsx')#error_bad_lines=False)
df.columns = list(range(0,len(df.columns)))
time = df.iloc[2].values[2:9]
df = df[3:-11]
```

```python
met_names = [df.iloc[i][1] for i in range(0,len(df.index),2)]
pyr_vals  = [df.iloc[i][2:9].values for i in range(0,len(df.index),2)]
pyr_sds   = [df.iloc[i+1][2:9].values for i in range(0,len(df.index),2)]
fru_vals  = [df.iloc[i][9:-1].values for i in range(0,len(df.index),2)]
fru_sds   = [df.iloc[i+1][9:-1].values for i in range(0,len(df.index),2)]
ctrl_vals = [df.iloc[i][-1:].values for i in range(0,len(df.index),2)]
ctrl_sds  = [df.iloc[i+1][-1:].values for i in range(0,len(df.index),2)]
mets_short = ['SUC','MAL','AKG','FUM','PEP','FBP','G6P','R5P','F6P', 'DHAP', '(I)CIT', 'AcCoA']
mets = ['Succinate','Malate','alpha ketoglutarate','fumarate','PEP','FBP','G6P','R5P','F6P', 'DHAP', 'citrate_isocitrate', 'AcCoA']
```

```python
step = [0, 1, 1, 1, 1, 0, 0, 0]
t_step = [0, 10, 15.0, 25.0, 40.0, 45.0, 55.0, 70.0]

font_size = 40
matplotlib.rc('xtick', labelsize=font_size) 
matplotlib.rc('ytick', labelsize=font_size)
for i,met in enumerate(mets):
    idx = met_names.index(met)
    plt.figure(figsize=(14,10))
    (_, caps, _) = plt.errorbar(time, pyr_vals[idx], yerr=pyr_sds[idx], linewidth=7, fmt='-o', markersize=15, capsize=30)
    plt.title(mets_short[i], fontsize=font_size+20)
    plt.xlabel('Time', fontsize=font_size)
#     plt.ylabel('Concentration', fontsize=font_size)
    plt.savefig('foo'+str(i)+'.png')
    for cap in caps:
        cap.set_markeredgewidth(4)
        
### plot control
### plot simulation

    
```
